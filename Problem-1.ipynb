```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

## uniform

```python
n_steps = np.arange(1001)
position = np.array([0])
ran = np.random.random((1000,))
ran = ran - 0.5

for i in range(1000):
    position = np.append(position, position[i] + ran[i])

plt.plot(n_steps, position)
plt.xlabel("n steps")
plt.ylabel("position")
```

```python
"""
steps: the steps each random walk take
n: calculate n samples of random walks

returns the final position of each random walk in 1d array
"""


def ran_walk_continue(steps, n):
    ans = np.zeros((n,))
    for i in range(n):
        ran = np.random.random((steps,))
        ran = ran - 0.5
        final_place = np.sum(ran)
        ans[i] = final_place
    return ans
```

```python
ans1 = ran_walk_continue(1000, 10000)
plt.hist(ans1, bins=50)
plt.show()
```

## discrete

```python
n_steps = np.arange(1001)
position = np.array([0])
ran = np.random.randint(2, size=1000)
ran = ran - 0.5

for i in range(1000):
    position = np.append(position, position[i] + ran[i])

plt.plot(n_steps, position)
plt.xlabel("n steps")
plt.ylabel("position")
```

```python
"""
steps: the steps each random walk take
n: calculate n samples of random walks

returns the final position of each random walk in 1d array
"""


def ran_walk_discrete(steps, n):
    ans = np.zeros((n,))
    for i in range(n):
        ran = np.random.randint(2, size=steps)
        ran = ran - 0.5
        final_place = np.sum(ran)
        ans[i] = final_place
    return ans
```
