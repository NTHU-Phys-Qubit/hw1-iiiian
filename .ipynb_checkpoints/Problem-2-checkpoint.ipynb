# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

# answer

### some pre defined func

```python
# position = [[x], [y]]
def r(position):
    ans = position[0]*position[1]
    return ans


#walk=[[x], [y]]
def trajectory(walk):
    position_x = np.array([0])
    position_y = np.array([0])
    
    for i in range(len(walk[0])):
        position_x = np.append(position_x, position_x[i] + walk[0][i])
        position_y = np.append(position_y, position_y[i] + walk[1][i])
    
    return [position_x, position_y]
```

### $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.

```python
ran_x = np.random.random((1000,)) - 0.5
ran_y = np.random.random((1000,)) - 0.5

position = trajectory([ran_x, ran_y])

plt.plot(position[0], position[1])
plt.xlabel("x")
plt.ylabel("y")
```

### a 2d random walk with a custom discrete step size on x and y axis

```python
ran_x = np.random.randint(2, size=1000)
ran_y = np.random.randint(2, size=1000)
ran_x = 2*(ran_x - 0.5)
ran_y = 2*(ran_y - 0.5)

position = trajectory([ran_x, ran_y])

plt.plot(position[0], position[1])
plt.xlabel("x")
plt.ylabel("y")
```

```python
"""
n_steps: the steps each random walk take
step_size: the size of a step on on x and y axis
n: calculate n samples of random walks

returns the final position of each random walk in array of shape (1,2), which is x and y position respectively
"""


def ran_step_2d(n_steps, step_size, n):
    ans_x = np.zeros((n,))
    ans_y = np.zeros((n,))
    for i in range(n):
        ran_x = np.random.randint(2, size=n_steps)
        ran_y = np.random.randint(2, size=n_steps)
        ran_x = 2 * step_size * (ran_x - 0.5)
        ran_y = 2 * step_size * (ran_y - 0.5)
        ans_x[i] = np.sum(ran_x)
        ans_y[i] = np.sum(ran_y)
    ans = np.array([ans_x, ans_y])
    return ans
```

```python
plt.hist(r(ans2), bins=50)
plt.show()
```

### $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

```python
ran = np.random.random((1000,))
ran = ran * 2 * np.pi

position = trajectory([np.cos(ran), np.sin(ran)])
    
plt.plot(position[0], position[1])
plt.xlabel("x")
plt.ylabel("y")
```

```python
"""
steps: the steps each random walk take
n: calculate n samples of random walks

returns the final position of each random walk in array of shape (1,2), which is x and y position respectively
"""


def rand_angle_2d(steps, n):
    ans_x = np.zeros((n,))
    ans_y = np.zeros((n,))

    for i in range(n):
        ran = np.random.random((steps,))
        ran = ran * 2 * np.pi
        x = np.cos(ran)
        y = np.sin(ran)
        ans_x[i] = np.sum(x)
        ans_y[i] = np.sum(y)

    ans = np.array([ans_x, ans_y])
    return ans
```

```python
ans3 = rand_angle_2d(1000, 10000)
```

```python
plt.hist2d(ans3[0], ans3[1], bins=50)
plt.colorbar()
plt.show()
```

```python
plt.hist(r(ans3), bins=50)
plt.show()
```
